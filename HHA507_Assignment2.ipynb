```python
# Check for missing values in a column
print(df.isnull().sum())
```

```python
# Identify and remove duplicate rows
print(df.duplicated())     #Identify
df.drop_duplicates(inplace=True)     #Remove Duplicate Rows
df
```

```python
df.duplicated().values.sum() # Checking if there are any duplicates
```

```python
# Remove leading and trailing white space from a column
df_columns = df.columns.str.strip()
df
```

```python
# Function to remove white space and special characters from a value
def clean_column(df):
  def clean_name(name):
    cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
    return cleaned_name.lower()
  df.columns = [clean_name(col) for col in df.columns]
  return df

# Apply the clean_value function to all columns
df = clean_column(df)
```

```python
# Check data types of columns
df.dtypes
```

```python
df = df[df["averagebp"] != 'missing']
df["averagebp"] = df["averagebp"].astype('int') # Changing the type to integter
```

```python
df['BpStatus'] = np.where(df['averagebp'] > 120, df['averagebp'], 'Above normal')
df['BpStatus'] = np.where(df['averagebp'] < 90, df['averagebp'], 'Below normal')
df
```

### Aggregating data using groupby and compute summary statistics

```python
df.groupby(['averagebp']).describe()    # Summary Statistics
```

```python
df = df.groupby('averagebp').agg({'sum'})    #Aggregate data
df
```
